```python
import pandas as pd
```

```python
long_message_chain_df = pd.read_csv('result100\LongMessageChain.csv')
long_message_chain_df.head()
```

```python
long_params_list_df = pd.read_csv('result100\LongParameterList.csv')
long_params_list_df.head()
```

```python
long_lambda_func_df = pd.read_csv('result100\LongLambdaFunction.csv')
long_lambda_func_df.head()
```

```python
def create_dataset_for_merging(df):
    smells = df['tuning machine'] == 1
    no_smells = df['tuning machine'] == 0

    df_smells = df[smells]
    df_no_smells = df[no_smells]
    random_df_no_smells = df_no_smells.sample(n=len(df_smells))

    final_df = pd.concat([df_smells, random_df_no_smells], ignore_index=True)
    final_df = final_df.sample(frac=1).reset_index(drop=True)
    final_df.drop(columns=['experience-based', 'statistics-based'], inplace=True)
    # final_df.head()
    return final_df
```

```python
LMC_df = create_dataset_for_merging(long_message_chain_df)
LPL_df = create_dataset_for_merging(long_params_list_df)
LLF_df = create_dataset_for_merging(long_lambda_func_df)
```

```python
column_mapping_LMC = {'subject':'Project', 'open':'File Name','lineno':'Line_LMC', 'LMC':'LMC', 'tuning machine':'LMCS'}
column_mapping_LPL = {'subject':'Project', 'open':'File Name','lineno':'Line_LPL', 'PAR':'LPLPAR', 'tuning machine':'LPL'}
column_mapping_LLF = {'subject':'Project', 'open':'File Name','lineno':'Line_LLF', 'PAR':'LLFPAR', 'tuning machine':'LLF'}

LMC_df.rename(columns=column_mapping_LMC, inplace=True)
LPL_df.rename(columns=column_mapping_LPL, inplace=True)
LLF_df.rename(columns=column_mapping_LLF, inplace=True)
```

```python
LMC_LPL = pd.merge(LMC_df, LPL_df, on=['Project', 'File Name'], how='outer')
main_df = pd.merge(LMC_LPL, LLF_df, on=['Project', 'File Name'], how='outer')
main_df.fillna(0, inplace=True)
main_df.info()
main_df.head(1000)
```

```python
main_df.to_csv('LMC_LPL_LLF.csv', index=False)
```
