```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from imblearn.over_sampling import SMOTE
%matplotlib inline 

df = pd.read_csv('code_smell_dataset.csv')
df
# df.shape
```

# Checking for Missing Values

```python
print(df.isnull().sum())
```

# Checking for Duplicated Values

```python
df.duplicated().sum()
df.loc[df.duplicated(), :]
```

```python
df.drop_duplicates(keep='first', inplace=True)
df.shape
```

# Visualizing the amount of code smells in a particular type

```python
sns.set_style('whitegrid')
sns.countplot(x='LC', data=df)
plt.show()
```

```python
sns.countplot(x='LM', data=df)
plt.show()
```

```python
sns.countplot(x='LMCS', data=df)
plt.show()
```

# Relationship between x_variables and y_variables using scatter plot

```python

sns.pairplot(df, x_vars=['CLOC','MLOC', 'LMC', 'LPLPAR','NOC','LLFPAR','NOO'], y_vars=['LC'])
sns.pairplot(df, x_vars=['CLOC','MLOC', 'LMC', 'LPLPAR','NOC','LLFPAR','NOO'], y_vars=['LM'])
sns.pairplot(df, x_vars=['CLOC','MLOC', 'LMC', 'LPLPAR','NOC','LLFPAR','NOO'], y_vars=['LMCS'])
sns.pairplot(df, x_vars=['CLOC','MLOC', 'LMC', 'LPLPAR','NOC','LLFPAR','NOO'], y_vars=['LPL'])
sns.pairplot(df, x_vars=['CLOC','MLOC', 'LMC', 'LPLPAR','NOC','LLFPAR','NOO'], y_vars=['LLF'])
plt.show()
```

# Correlation between the variables

```python
corr = df[['CLOC','MLOC', 'LMC', 'LPLPAR','NOC','LLFPAR','NOO','LC', 'LM', 'LMCS', 'LPL', 'LLF']].corr()
sns.heatmap(corr, annot=True)
plt.rcParams['figure.figsize'] = (17,12)
plt.show()
```
