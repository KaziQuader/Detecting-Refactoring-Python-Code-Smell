# Model Implemenatation

```python
# Label Power set GBBoost
from sklearn.model_selection import RandomizedSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from scipy.stats import randint, uniform
from skmultilearn.problem_transform import LabelPowerset
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score
from sklearn.metrics import hamming_loss


x = df[['CLOC', 'MLOC', 'LMC', 'LPLPAR', 'NOC', 'LLFPAR', 'NOO']]
y = df[['LC', 'LM', 'LMCS', 'LPL', 'LLF']]

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42, stratify=y)

param_dist = {
    'classifier__n_estimators': randint(50, 200),
    'classifier__learning_rate': uniform(0.01, 0.3),
    'classifier__max_depth': randint(3, 10),
    'classifier__subsample': uniform(0.6, 0.4),
    'classifier__min_samples_split': randint(2, 20),
    'classifier__min_samples_leaf': randint(1, 20),
}

# Initialize the LabelPowerset classifier with a Decision Tree base classifier
base_classifier = GradientBoostingClassifier()
classifier = LabelPowerset(classifier=base_classifier)
random_search = RandomizedSearchCV(
    classifier,
    param_distributions=param_dist,
    n_iter=10,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)


# Train the LabelPowerset classifier
random_search.fit(X_train, y_train)

print("Best Parameters: ", random_search.best_params_)
print("Best Score: ", random_search.best_score_)


# Make predictions on the test set
best_model = random_search.best_estimator_
predictions = best_model.predict(X_test)

classes = random_search.best_estimator_.classifier_.classes_
print("Classes:", classes)

# Evaluate the performance
accuracy = accuracy_score(y_test, predictions)
report = classification_report(y_test, predictions)
# accuracy = accuracy_score(y_test, predictions)
f1 = f1_score(y_test, predictions, average='micro')
precision = precision_score(y_test, predictions, average='micro')
recall = recall_score(y_test, predictions, average='micro')
hamming_loss_value = hamming_loss(y_test, predictions)

# Print the results
print(f'Accuracy: {accuracy}')
print(f'F1: {f1}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print('Classification Report:')
print(f'Hamming Loss: {hamming_loss_value}')
print(report)
```
